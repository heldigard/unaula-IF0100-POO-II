## 1. Variables y Tipos de Datos

** Definición:** Una **variable** es un nombre que referencia a un valor almacenado en memoria.

** Tipos básicos en Python:**

| Tipo | Descripción | Ejemplo |
|------|-------------|----------|
| `int` | Enteros | `42`, `-10`, `0` |
| `float` | Decimales | `3.14`, `-0.5`, `2.0` |
| `str` | Texto (cadena) | `"Hola"`, `'Mundo'` |
| `bool` | Booleanos | `True`, `False` |

**✅ Ejemplo:**

````python
# Declaración de variables
nombre = "Ana"           # str (string)
edad = 25                # int (integer)
altura = 1.68            # float
es_estudiante = True     # bool (boolean)

# Verificar el tipo de una variable
print(f"Tipo de nombre: {type(nombre)}")
print(f"Tipo de edad: {type(edad)}")
print(f"Tipo de altura: {type(altura)}")
print(f"Tipo de es_estudiante: {type(es_estudiante)}")

# Mostrar valores
print(f"\nHola, me llamo {nombre}, tengo {edad} años y mido {altura}m")
````

** Nota importante:** En Python NO necesitas declarar el tipo. El tipo se infiere del valor asignado.

###  Reglas para nombres de variables

✅ **Válidos:**
- `nombre`
- `_contador`
- `edad1`
- `valorMaximo`

❌ **NO válidos:**
- `1nombre` (no empieza con número)
- `mi-variable` (no usa guiones, usa `_`)
- `class` (palabra reservada)
- `mi variable` (no tiene espacios)

## 2. Operadores Aritméticos

** Operadores aritméticos:**

| Operador | Descripción | Ejemplo |
|----------|-------------|----------|
| `+` | Suma | `5 + 3 = 8` |
| `-` | Resta | `5 - 3 = 2` |
| `*` | Multiplicación | `5 * 3 = 15` |
| `/` | División (float) | `10 / 3 = 3.333...` |
| `//` | División entera | `10 // 3 = 3` |
| `%` | Módulo (resto) | `10 % 3 = 1` |
| `**` | Potencia | `2 ** 3 = 8` |

**✅ Ejemplo:**

````python
a = 17
b = 5

print(f"Suma: {a} + {b} = {a + b}")
print(f"Resta: {a} - {b} = {a - b}")
print(f"Multiplicación: {a} * {b} = {a * b}")
print(f"División: {a} / {b} = {a / b}")
print(f"División entera: {a} // {b} = {a // b}")
print(f"Módulo: {a} % {b} = {a % b}")
print(f"Potencia: {a} ** {b} = {a ** b}")
````

** ¿Para qu e sirve el módulo `%`?**

El módulo es útil para:
- Verificar si un número es par: `x % 2 == 0`
- Ciclos que alternan valores
- Obtener el d d7gito más a la derecha de un número

## 3. Operadores de Comparación

** Operadores de comparación:**

| Operador | Descripción | Ejemplo |
|----------|-------------|----------|
| `==` | Igual a | `5 == 5` → `True` |
| `!=` | Diferente de | `5 != 3` → `True` |
| `<` | Menor que | `3 < 5` → `True` |
| `>` | Mayor que | `5 > 3` → `True` |
| `<=` | Menor o igual | `3 <= 3` → `True` |
| `>=` | Mayor o igual | `5 >= 5` → `True` |

**✅ Ejemplo:**

````python
edad = 18

print(f"edad == 18: {edad == 18}")
print(f"edad != 18: {edad != 18}")
print(f"edad < 18: {edad < 18}")
print(f"edad > 18: {edad > 18}")
print(f"edad <= 18: {edad <= 18}")
print(f"edad >= 18: {edad >= 18}")
````

**⚠️ Importante:** 
- `=` es **asignación** (`x = 5`)
- `==` es **comparación** (`x == 5`)

## 4. Entrada y Salida de Datos

** `print()` - Mostrar información:**

```python
print("Hola mundo")              # Texto simple
print("Valor:", x)              # Múltiples valores
print(f"Valor: {x}")            # f-string (interpolación)
```

** `input()` - Recibir información:**

```python
nombre = input("Tu nombre: ")    # Siempre devuelve string
edad = int(input("Tu edad: "))  # Convertir a int
```

**✅ Ejemplo:

````python
# Ejemplo de print con diferentes formatos
producto = "Laptop"
precio = 1500

# Forma antigua (no recomendada)
print("El producto es " + producto + " y cuesta " + str(precio))

# Forma con .format() (mejor)
print("El producto es {} y cuesta {}".format(producto, precio))

# Forma con f-string ( a1RECOMENDADA!)
print(f"El producto es {producto} y cuesta ${precio}")
````

** Conversión de tipos (casting):**

| Función | Descripción |
|----------|----------|
| `int(x)` | Convierte a entero |
| `float(x)` | Convierte a decimal |
| `str(x)` | Convierte a texto |
| `bool(x)` | Convierte a booleano |

### ✅ Validación de la Solución

````python
# Tests automatizados para el ejercicio
def verificar_calculadora():
    """Verifica que la calculadora esté correctamente implementada."""
    # Simular entrada del usuario
    test_cases = [
        (10, 5, "+", 15),
        (10, 5, "-", 5),
        (10, 5, "*", 50),
        (10, 5, "/", 2),
        (7, 3, "%", 1),
        (2, 3, "**", 8),
    ]
    
    for num1, num2, op, esperado in test_cases:
        # Calcular resultado
        if op == "+":
            resultado = num1 + num2
        elif op == "-":
            resultado = num1 - num2
        elif op == "*":
            resultado = num1 * num2
        elif op == "/":
            resultado = num1 / num2
        elif op == "%":
            resultado = num1 % num2
        elif op == "**":
            resultado = num1 ** num2
        else:
            print(f"❌ Operación no válida: {op}")
            return
        
        assert resultado == esperado, f"Error: {num1} {op} {num2} = {resultado}, esperaba {esperado}"
        print(f"✅ {num1} {op} {num2} = {resultado}")
    
    print("\n ¡Excelente! Todos los tests pasaron!")

# Ejecutar validación
verificar_calculadora()
````
