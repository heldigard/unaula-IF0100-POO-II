## üîÄ Condicionales: if, elif, else

````python
# ============================================
# CONDICIONAL IF B√ÅSICO
# ============================================

edad = 18

if edad >= 18:
    print("Eres mayor de edad")

print("Fin del programa")  # Esta l√≠nea siempre se ejecuta
````

````python
# ============================================
# IF - ELSE
# ============================================

nota = 3.5

if nota >= 3.0:
    print("¬°Aprobado! üéâ")
else:
    print("Reprobado üò¢")

print(f"Tu nota fue: {nota}")
````

````python
# ============================================
# IF - ELIF - ELSE
# ============================================

calificacion = 85  # De 0 a 100

if calificacion >= 90:
    letra = "A"
    mensaje = "Excelente"
elif calificacion >= 80:
    letra = "B"
    mensaje = "Muy bien"
elif calificacion >= 70:
    letra = "C"
    mensaje = "Bien"
elif calificacion >= 60:
    letra = "D"
    mensaje = "Regular"
else:
    letra = "F"
    mensaje = "Necesitas mejorar"

print(f"Calificaci√≥n: {calificacion}")
print(f"Letra: {letra}")
print(f"Mensaje: {mensaje}")
````

````python
# ============================================
# CONDICIONALES ANIDADOS
# ============================================

usuario_registrado = True
contrasena_correcta = True
tiene_permiso = False

if usuario_registrado:
    print("Usuario existe en el sistema")
    
    if contrasena_correcta:
        print("Contrase√±a correcta")
        
        if tiene_permiso:
            print("‚úÖ Acceso concedido")
        else:
            print("‚ùå No tienes permisos")
    else:
        print("‚ùå Contrase√±a incorrecta")
else:
    print("‚ùå Usuario no registrado")
````

````python
# ============================================
# OPERADORES L√ìGICOS EN CONDICIONALES
# ============================================

# AND - Ambas condiciones deben ser verdaderas
edad = 25
tiene_licencia = True

if edad >= 18 and tiene_licencia:
    print("Puede conducir")
else:
    print("No puede conducir")

# OR - Al menos una condici√≥n debe ser verdadera
es_estudiante = True
es_profesor = False

if es_estudiante or es_profesor:
    print("Tiene acceso a la plataforma")

# NOT - Niega la condici√≥n
esta_bloqueado = False

if not esta_bloqueado:
    print("Cuenta activa")
````

````python
# ============================================
# OPERADOR TERNARIO (EXPRESI√ìN CONDICIONAL)
# ============================================

edad = 20

# Forma larga
if edad >= 18:
    estado = "Mayor de edad"
else:
    estado = "Menor de edad"

# Forma corta (ternaria)
estado = "Mayor de edad" if edad >= 18 else "Menor de edad"

print(estado)

# Ejemplo m√°s complejo
nota = 4.2
resultado = "Aprobado" if nota >= 3.0 else "Reprobado"
print(f"Resultado: {resultado}")
````

## üîÅ Bucles: for

````python
# ============================================
# BUCLE FOR CON RANGE()
# ============================================

# range(n) -> 0, 1, 2, ..., n-1
print("Range(5):")
for i in range(5):
    print(f"  Iteraci√≥n {i}")

# range(inicio, fin) -> inicio, ..., fin-1
print("\nRange(2, 6):")
for i in range(2, 6):
    print(f"  Valor: {i}")

# range(inicio, fin, paso)
print("\nRange(0, 10, 2) - Pares:")
for i in range(0, 10, 2):
    print(f"  {i}")

# Range descendente
print("\nRange(5, 0, -1) - Cuenta regresiva:")
for i in range(5, 0, -1):
    print(f"  {i}...")
print("¬°Despegue! üöÄ")
````

````python
# ============================================
# FOR CON LISTAS
# ============================================

frutas = ["manzana", "banana", "cereza", "durazno"]

print("Lista de frutas:")
for fruta in frutas:
    print(f"  üçé {fruta}")

# Con enumerate() - √≠ndice y valor
print("\nCon √≠ndice:")
for indice, fruta in enumerate(frutas):
    print(f"  {indice + 1}. {fruta}")

# Con enumerate() y start
print("\nEnumerate(start=1):")
for num, fruta in enumerate(frutas, start=1):
    print(f"  {num}. {fruta}")
````

````python
# ============================================
# FOR CON STRINGS
# ============================================

mensaje = "Python"

print("Caracteres del mensaje:")
for letra in mensaje:
    print(f"  {letra}")

# Contar vocales
texto = "Programaci√≥n Orientada a Objetos"
vocales = "aeiouAEIOU"
contador = 0

for letra in texto:
    if letra in vocales:
        contador += 1

print(f"\nEn '{texto}'")
print(f"Hay {contador} vocales")
````

## üîÅ Bucles: while

````python
# ============================================
# BUCLE WHILE B√ÅSICO
# ============================================

contador = 0

while contador < 5:
    print(f"Contador: {contador}")
    contador += 1  # ¬°Importante! Evita bucle infinito

print("Fin del while")
````

````python
# ============================================
# WHILE CON CONDICI√ìN COMPLEJA
# ============================================

# Simulaci√≥n: Intentos de login
intentos = 3
contrasena_correcta = "python123"
acceso_concedido = False

while intentos > 0 and not acceso_concedido:
    print(f"\nIntentos restantes: {intentos}")
    # En un programa real: contrasena = input("Contrase√±a: ")
    contrasena = "python123"  # Simulaci√≥n
    
    if contrasena == contrasena_correcta:
        acceso_concedido = True
        print("‚úÖ Acceso concedido")
    else:
        intentos -= 1
        print("‚ùå Contrase√±a incorrecta")

if not acceso_concedido:
    print("\nüö´ Cuenta bloqueada. Contacte al administrador.")
````

## üõë Control de Bucles: break, continue, pass

````python
# ============================================
# BREAK - Salir del bucle
# ============================================

print("Buscando el n√∫mero 7:")
for num in range(1, 20):
    if num == 7:
        print(f"  ¬°Encontrado en {num}!")
        break  # Sale del bucle inmediatamente
    print(f"  {num} no es...")

print("Bucle terminado")
````

````python
# ============================================
# CONTINUE - Saltar a la siguiente iteraci√≥n
# ============================================

print("N√∫meros impares del 1 al 10:")
for num in range(1, 11):
    if num % 2 == 0:  # Si es par
        continue      # Salta a la siguiente iteraci√≥n
    print(f"  {num}")  # Solo se ejecuta para impares
````

````python
# ============================================
# PASS - No hace nada (placeholder)
# ============================================

# √ötil cuando necesitas sintaxis v√°lida pero a√∫n no implementas

for i in range(5):
    if i == 2:
        pass  # TODO: Implementar l√≥gica especial
    else:
        print(f"Procesando {i}")

# Tambi√©n usado para clases/funciones vac√≠as
def funcion_pendiente():
    pass  # Por implementar

class MiClase:
    pass  # Por implementar
````

## ‚ö° Comprensiones de Lista

````python
# ============================================
# LIST COMPREHENSION
# ============================================

# Forma tradicional
cuadrados = []
for x in range(10):
    cuadrados.append(x ** 2)
print(f"Tradicional: {cuadrados}")

# Forma con comprensi√≥n
cuadrados = [x ** 2 for x in range(10)]
print(f"Comprensi√≥n: {cuadrados}")

# Con condici√≥n
pares = [x for x in range(20) if x % 2 == 0]
print(f"Pares: {pares}")

# Con if-else
categorias = ["par" if x % 2 == 0 else "impar" for x in range(10)]
print(f"Categor√≠as: {categorias}")
````

## üîó Conexi√≥n con TaskFlow

En **TaskFlow**, las estructuras de control son fundamentales:

```python
# Ejemplo: Validaci√≥n de tareas
tareas = [
    {"titulo": "Dise√±ar DB", "estado": "completada", "prioridad": 1},
    {"titulo": "Crear API", "estado": "en_progreso", "prioridad": 1},
    {"titulo": "Tests", "estado": "pendiente", "prioridad": 2},
]

# Filtrar tareas urgentes no completadas
print("Tareas urgentes pendientes:")
for tarea in tareas:
    if tarea["prioridad"] == 1 and tarea["estado"] != "completada":
        print(f"  ‚ö†Ô∏è {tarea['titulo']} - {tarea['estado']}")

# Contar tareas completadas
completadas = sum(1 for t in tareas if t["estado"] == "completada")
print(f"\nProgreso: {completadas}/{len(tareas)} tareas completadas")
```
