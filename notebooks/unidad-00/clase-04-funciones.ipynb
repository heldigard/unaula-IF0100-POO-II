## ðŸ”§ DefiniciÃ³n de Funciones

````python
# ============================================
# FUNCIÃ“N BÃSICA
# ============================================

# DefiniciÃ³n
def saludar():
    """Muestra un saludo en pantalla."""
    print("Â¡Hola, bienvenido al curso!")

# Llamada a la funciÃ³n
saludar()
saludar()  # Se puede llamar mÃºltiples veces
````

````python
# ============================================
# FUNCIÃ“N CON PARÃMETROS
# ============================================

def saludar_usuario(nombre):
    """Saluda a un usuario especÃ­fico."""
    print(f"Â¡Hola, {nombre}! Â¿CÃ³mo estÃ¡s?")

# Llamadas con diferentes argumentos
saludar_usuario("Ana")
saludar_usuario("Carlos")

# MÃºltiples parÃ¡metros
def presentar(nombre, edad, ciudad):
    """Presenta a una persona."""
    print(f"Me llamo {nombre}, tengo {edad} aÃ±os y vivo en {ciudad}.")

presentar("MarÃ­a", 25, "MedellÃ­n")

# Argumentos nombrados (mÃ¡s legible)
presentar(ciudad="BogotÃ¡", nombre="Luis", edad=30)
````

````python
# ============================================
# FUNCIÃ“N CON VALOR DE RETORNO
# ============================================

def sumar(a, b):
    """Retorna la suma de dos nÃºmeros."""
    resultado = a + b
    return resultado

# Capturar el retorno
total = sumar(5, 3)
print(f"5 + 3 = {total}")

# Usar directamente
print(f"10 + 20 = {sumar(10, 20)}")

# MÃºltiples valores de retorno
def operaciones(a, b):
    """Retorna suma, resta, multiplicaciÃ³n y divisiÃ³n."""
    return a + b, a - b, a * b, a / b

suma, resta, mult, div = operaciones(10, 3)
print(f"Suma: {suma}, Resta: {resta}, Mult: {mult}, Div: {div:.2f}")
````

## âš™ï¸ ParÃ¡metros Avanzados

````python
# ============================================
# PARÃMETROS POR DEFECTO
# ============================================

def crear_usuario(nombre, rol="estudiante", activo=True):
    """Crea un diccionario de usuario."""
    return {
        "nombre": nombre,
        "rol": rol,
        "activo": activo
    }

# Usar valores por defecto
usuario1 = crear_usuario("Juan")
print(f"Por defecto: {usuario1}")

# Sobrescribir algunos
usuario2 = crear_usuario("Ana", rol="profesor")
print(f"Rol cambiado: {usuario2}")

# Todos personalizados
usuario3 = crear_usuario("Luis", "admin", False)
print(f"Personalizado: {usuario3}")
````

````python
# ============================================
# *ARGS - ARGUMENTOS VARIABLES
# ============================================

def sumar_todos(*numeros):
    """Suma todos los nÃºmeros recibidos."""
    total = 0
    for num in numeros:
        total += num
    return total

# Diferentes cantidades de argumentos
print(f"Sumar 1 nÃºmero: {sumar_todos(5)}")
print(f"Sumar 3 nÃºmeros: {sumar_todos(1, 2, 3)}")
print(f"Sumar 5 nÃºmeros: {sumar_todos(10, 20, 30, 40, 50)}")

# args es una tupla
def mostrar_argumentos(*args):
    print(f"Tipo: {type(args)}")
    print(f"Argumentos: {args}")
    for i, arg in enumerate(args):
        print(f"  args[{i}] = {arg}")

mostrar_argumentos(1, "hola", True, 3.14)
````

````python
# ============================================
# **KWARGS - ARGUMENTOS NOMBRADOS VARIABLES
# ============================================

def crear_perfil(**kwargs):
    """Crea un perfil con datos variables."""
    print("Perfil creado:")
    for clave, valor in kwargs.items():
        print(f"  {clave}: {valor}")
    return kwargs

perfil = crear_perfil(
    nombre="Carlos",
    edad=28,
    ciudad="Cali",
    profesion="Ingeniero"
)

# Combinar *args y **kwargs
def funcion_completa(*args, **kwargs):
    print(f"Args posicionales: {args}")
    print(f"Args nombrados: {kwargs}")

funcion_completa(1, 2, 3, nombre="Ana", edad=25)
````

## ðŸ“¦ Scope de Variables

````python
# ============================================
# ALCANCE (SCOPE) DE VARIABLES
# ============================================

# Variable global
contador_global = 0

def incrementar():
    """Intenta modificar variable global."""
    # Esto crearÃ¡ una variable local, no modifica la global
    contador_global = 10  # âš ï¸ Variable local!
    print(f"Dentro de funciÃ³n: {contador_global}")

incrementar()
print(f"Fuera de funciÃ³n: {contador_global}")  # Sigue siendo 0

# Para modificar global, usa 'global'
def incrementar_global():
    """Modifica variable global."""
    global contador_global
    contador_global += 1
    print(f"Dentro: {contador_global}")

incrementar_global()
print(f"Fuera: {contador_global}")
````

## âš¡ Funciones Lambda

````python
# ============================================
# FUNCIONES LAMBDA (ANÃ“NIMAS)
# ============================================

# FunciÃ³n normal
def cuadrado_normal(x):
    return x ** 2

# Equivalente lambda
cuadrado = lambda x: x ** 2

print(f"Normal: {cuadrado_normal(5)}")
print(f"Lambda: {cuadrado(5)}")

# Lambda con mÃºltiples parÃ¡metros
suma = lambda a, b: a + b
print(f"Suma: {suma(3, 4)}")

# Lambda en funciones de orden superior
numeros = [1, 2, 3, 4, 5]

# map() con lambda
cuadrados = list(map(lambda x: x ** 2, numeros))
print(f"Cuadrados: {cuadrados}")

# filter() con lambda
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(f"Pares: {pares}")

# sorted() con lambda
estudiantes = [
    {"nombre": "Ana", "nota": 4.5},
    {"nombre": "Luis", "nota": 3.8},
    {"nombre": "MarÃ­a", "nota": 4.9}
]

ordenados = sorted(estudiantes, key=lambda e: e["nota"], reverse=True)
print(f"Ordenados por nota: {ordenados}")
````
