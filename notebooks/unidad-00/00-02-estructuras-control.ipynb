## 1. Condicionales: if / elif / else

**🗂 Definicin:** Los condicionales permiten ejecutar diferentes bloques de cdigo seg fan condiciones.

**📋 ¡Importante!** En Python, la **indentacin** define los bloques de cdigo (4 espacios).

**✅ Sintaxis:**

```python
if condicion:
    # Cdigo si condicin es True
elif otra_condicion:
    # Cdigo si otra_condicin es True
else:
    # Cdigo si ninguna condicin anterior es True
```

````python
# Ejemplo 1: Clasificar edad
edad = 17

if edad < 13:
    print("👶 Ni f1o")
elif edad < 18:
    print("👦 Adolescente")
elif edad < 65:
    print("👨 Adulto")
else:
    print("👵 Adulto mayor")
````

````python
# Ejemplo 2: Calcular nota cualitativa
nota = 85

if nota >= 90:
    calificacion = "A - 
1Excelente!"
elif nota >= 80:
    calificacion = "B - Muy bien"
elif nota >= 70:
    calificacion = "C - Bien"
elif nota >= 60:
    calificacion = "D - Suficiente"
else:
    calificacion = "F - Reprobado"

print(f"Nota: {nota} - {calificacion}")
````

````python
# Ejemplo 3: M faltiples condiciones con AND/OR
edad = 25
tiene_licencia = True

# AND: ambas condiciones deben ser True
if edad >= 18 and tiene_licencia:
    print("🚙 Puede conducir")
else:
    print("🚫 No puede conducir")

# OR: al menos una condicin debe ser True
dia = "s e1bado"
if dia == "s e1bado" or dia == "domingo":
    print("🌟 Es fin de semana")
else:
    print("💼 Es d da laboral")
````

## 2. Ciclo FOR

**🗂 Definicin:** El ciclo `for` se usa para iterar sobre secuencias (listas, strings, rangos).

**✅ Sintaxis:**

```python
for variable in secuencia:
    # Cdigo a repetir
```

````python
# Ejemplo 1: Iterar sobre un string
mensaje = "Python"

for letra in mensaje:
    print(f"Letra: {letra}")
````

````python
# Ejemplo 2: Usar range()
print("Contar del 0 al 4:")
for i in range(5):
    print(i)

print("\nContar del 1 al 5:")
for i in range(1, 6):
    print(i)

print("\nContar de 2 en 2:")
for i in range(0, 10, 2):
    print(i)
````

````python
# Ejemplo 3: Tabla de multiplicar
numero = 7

print(f"Tabla del {numero}:")
for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
````

````python
# Ejemplo 4: Sumar elementos de una lista
numeros = [10, 20, 30, 40, 50]
suma = 0

for num in numeros:
    suma += num  # Equivale a: suma = suma + num

print(f"La suma es: {suma}")
print(f"El promedio es: {suma / len(numeros)}")
````

### 📋 Funcin range()

| Sintaxis | Descripcin |
|----------|----------|
| `range(n)` | Del 0 hasta n-1 |
| `range(a, b)` | De a hasta b-1 |
| `range(a, b, paso)` | De a hasta b-1 con el paso indicado |

## 3. Ciclo WHILE

**🗂 Definicin:** El ciclo `while` repite un bloque de cdigo mientras una condicin sea verdadera.

**⚠️ Cuidado:** Si la condicin nunca se vuelve falsa, se crea un **bucle infinito**.

**✅ Sintaxis:**

```python
while condicion:
    # Cdigo a repetir
```

````python
# Ejemplo 1: Contador regresivo
contador = 5

while contador > 0:
    print(f"⏰ Contador: {contador}")
    contador -= 1  # Equivale a: contador = contador - 1

print("💥 
1Despegue!")
````

## 4. Break y Continue

**🗂 Definicin:**
- `break`: Termina el ciclo inmediatamente
- `continue`: Salta a la siguiente iteracin

**✅ Ejemplo:

````python
# Ejemplo con BREAK
print("Buscar el n famero 7:")
for i in range(1, 11):
    if i == 7:
        print(f"✅ 
1Encontrado! El n famero es {i}")
        break
    print(f"Buscando... {i}")
````

````python
# Ejemplo con CONTINUE
print("\nImprimir sdlo n fameros pares:")
for i in range(1, 11):
    if i % 2 != 0:
        continue  # Saltar si es impar
    print(f"Par: {i}")
````

````python
# Ejemplo combinado
print("\nProcesar n fameros (ignorar negativos, detener en 100):")
numeros = [10, -5, 20, -3, 100, 30, 50]

for num in numeros:
    if num < 0:
        print(f"\u274c Ignorando {num} (negativo)")
        continue
    if num == 100:
        print(f"🚫 
1L admite encontrado! Deteniendo...")
        break
    print(f"Procesando {num}")
````
