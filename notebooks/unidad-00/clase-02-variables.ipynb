## ðŸ“š TeorÃ­a: Variables en Python

### Â¿QuÃ© es una Variable?

Una variable es un **nombre simbÃ³lico** que referencia un valor almacenado en memoria.

**CaracterÃ­sticas de Python:**
- No necesitas declarar el tipo (tipado dinÃ¡mico)
- Usa snake_case por convenciÃ³n
- Son case-sensitive (edad â‰  Edad â‰  EDAD)

### Reglas para Nombres de Variables

âœ… **Correcto:**
- `nombre_usuario`
- `_privado`
- `contador1`
- `MAXIMO_INTENTOS` (constantes)

âŒ **Incorrecto:**
- `1contador` (no puede empezar con nÃºmero)
- `nombre-usuario` (guiones no permitidos)
- `class` (palabra reservada)
- `nombre usuario` (espacios no permitidos)

## ðŸ”¢ Tipos de Datos BÃ¡sicos

````python
# ============================================
# TIPOS NUMÃ‰RICOS
# ============================================

# int - NÃºmeros enteros
edad = 25
cantidad_tareas = 10
temperatura = -5

print(f"Edad: {edad} - Tipo: {type(edad).__name__}")

# float - NÃºmeros decimales
precio = 29.99
pi = 3.14159
nota_promedio = 4.5

print(f"Precio: {precio} - Tipo: {type(precio).__name__}")

# complex - NÃºmeros complejos
numero_complejo = 3 + 4j
print(f"Complejo: {numero_complejo} - Tipo: {type(numero_complejo).__name__}")
````

````python
# ============================================
# TIPO STRING (str)
# ============================================

# DefiniciÃ³n de strings
nombre = 'Juan'                    # Comillas simples
apellido = "PÃ©rez"                 # Comillas dobles
descripcion = '''Texto largo       # Triples comillas
que puede ocupar                    # para textos multilÃ­nea
varias lÃ­neas'''

print(f"Nombre: {nombre} - Tipo: {type(nombre).__name__}")

# Operaciones comunes con strings
saludo = "Hola"
nombre_completo = saludo + " " + nombre  # ConcatenaciÃ³n
repetido = "Ha" * 3                       # RepeticiÃ³n

print(f"ConcatenaciÃ³n: {nombre_completo}")
print(f"RepeticiÃ³n: {repetido}")

# f-strings (formateo moderno)
edad = 25
mensaje = f"{nombre} tiene {edad} aÃ±os y en 5 aÃ±os tendrÃ¡ {edad + 5}"
print(mensaje)
````

````python
# ============================================
# TIPO BOOLEANO (bool)
# ============================================

# Valores booleanos
es_activo = True
es_admin = False
tarea_completada = True

print(f"Es activo: {es_activo} - Tipo: {type(es_activo).__name__}")

# Operaciones lÃ³gicas
resultado_and = True and False   # False
resultado_or = True or False     # True
resultado_not = not True         # False

print(f"AND: {resultado_and}, OR: {resultado_or}, NOT: {resultado_not}")

# Booleanos en contextos numÃ©ricos
print(f"True como nÃºmero: {int(True)}")   # 1
print(f"False como nÃºmero: {int(False)}") # 0
````

````python
# ============================================
# TIPO None (NoneType)
# ============================================

# None representa la ausencia de valor
resultado = None
proxima_tarea = None
usuario_actual = None

print(f"Resultado: {resultado} - Tipo: {type(resultado).__name__}")

# Verificar si es None
if resultado is None:
    print("No hay resultado aÃºn")

# ComparaciÃ³n con == vs is
print(f"Â¿resultado == None? {resultado == None}")  # Funciona pero no recomendado
print(f"Â¿resultado is None? {resultado is None}")  # âœ… Forma correcta
````

## ðŸ”„ ConversiÃ³n de Tipos (Casting)

````python
# ============================================
# CONVERSIÃ“N ENTRE TIPOS
# ============================================

# int() - Convertir a entero
numero_str = "42"
numero_int = int(numero_str)
print(f"'{numero_str}' -> {numero_int} (tipo: {type(numero_int).__name__})")

# float() - Convertir a decimal
precio_str = "19.99"
precio_float = float(precio_str)
print(f"'{precio_str}' -> {precio_float} (tipo: {type(precio_float).__name__})")

# str() - Convertir a string
cantidad = 100
cantidad_str = str(cantidad)
print(f"{cantidad} -> '{cantidad_str}' (tipo: {type(cantidad_str).__name__})")

# bool() - Convertir a booleano
print(f"bool(1) = {bool(1)}")      # True
print(f"bool(0) = {bool(0)}")      # False
print(f"bool('') = {bool('')}")    # False
print(f"bool('a') = {bool('a')}")  # True
print(f"bool([]) = {bool([])}")    # False
print(f"bool([1]) = {bool([1])}")  # True
````

## âž• Operadores

````python
# ============================================
# OPERADORES ARITMÃ‰TICOS
# ============================================

a = 17
b = 5

print(f"a = {a}, b = {b}")
print(f"Suma (a + b): {a + b}")           # 22
print(f"Resta (a - b): {a - b}")          # 12
print(f"MultiplicaciÃ³n (a * b): {a * b}") # 85
print(f"DivisiÃ³n (a / b): {a / b}")       # 3.4
print(f"DivisiÃ³n entera (a // b): {a // b}")  # 3
print(f"MÃ³dulo (a % b): {a % b}")         # 2
print(f"Potencia (a ** b): {a ** b}")     # 1419857
````

````python
# ============================================
# OPERADORES DE COMPARACIÃ“N
# ============================================

x = 10
y = 20

print(f"x = {x}, y = {y}")
print(f"Igual (x == y): {x == y}")        # False
print(f"Diferente (x != y): {x != y}")    # True
print(f"Mayor (x > y): {x > y}")          # False
print(f"Menor (x < y): {x < y}")          # True
print(f"Mayor o igual (x >= y): {x >= y}")  # False
print(f"Menor o igual (x <= y): {x <= y}")  # True

# ComparaciÃ³n de strings
print(f"\n'a' < 'b': {'a' < 'b'}")  # True (orden alfabÃ©tico)
````

````python
# ============================================
# OPERADORES DE ASIGNACIÃ“N
# ============================================

contador = 10
print(f"Valor inicial: {contador}")

contador += 5   # contador = contador + 5
print(f"DespuÃ©s de += 5: {contador}")

contador -= 3   # contador = contador - 3
print(f"DespuÃ©s de -= 3: {contador}")

contador *= 2   # contador = contador * 2
print(f"DespuÃ©s de *= 2: {contador}")

contador /= 4   # contador = contador / 4
print(f"DespuÃ©s de /= 4: {contador}")

# TambiÃ©n existen: //=, %=, **=
````
