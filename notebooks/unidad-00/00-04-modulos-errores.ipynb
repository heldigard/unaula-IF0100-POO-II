## 1. Mdulos en Python

**🗂 Definicin:** Un **mdulo** es un archivo `.py` que contiene definiciones y declaraciones de Python.

**📋 Mdulos est e1ndar de la biblioteca:**
- `math` - Funciones matem e1ticas
- `random` - Generaci f3n de n fameros aleatorios
- `datetime` - Fechas y horas
- `json` - Codificaci f3n/decodificaci f3n JSON
- `os` - Funciones del sistema operativo
- `pathlib` - Rutas de archivos

````python
# 📌 Mdulo math
import math

print(f"Pi: {math.pi}")
print(f"Euler: {math.e}")
print(f"Ra d7z de 25: {math.sqrt(25)}")
print(f"Potencia: 2^3 = {math.pow(2, 3)}")
print(f"Sen(90 b0): {math.sin(math.pi/2):.2f}")
print(f"Redondear hacia arriba: {math.ceil(4.2)}")
print(f"Redondear hacia abajo: {math.floor(4.8)}")
````

````python
# 🔀 Mdulo random
import random

print(f"Entero aleatorio (1-10): {random.randint(1, 10)}")
print(f"Flotante aleatorio (0-1): {random.random():.4f}")
print(f"Opci f3n aleatoria: {random.choice(['rojo', 'verde', 'azul'])}")

# Mezclar lista
cartas = ['A', 'K', 'Q', 'J', '10']
random.shuffle(cartas)
print(f"Cartas mezcladas: {cartas}")

# Muestra aleatoria
numeros = list(range(1, 11))
muestra = random.sample(numeros, 3)
print(f"Muestra de 3: {muestra}")
````

````python
# 🕒 Mdulo datetime
from datetime import datetime, date, timedelta

# Fecha y hora actual
ahora = datetime.now()
print(f"Fecha y hora: {ahora}")
print(f"A f1o: {ahora.year}")
print(f"Mes: {ahora.month}")
print(f"D d7a: {ahora.day}")

# Formatear fecha
print(f"\nFormateada: {ahora.strftime('%d/%m/%Y %H:%M')}")

# Crear fecha espec d7fica
navidad = date(2026, 12, 25)
hoy = date.today()
dias_para_navidad = (navidad - hoy).days
print(f"\nD d7as hasta Navidad: {dias_para_navidad}")

# Sumar d d7as
ma f1ana = hoy + timedelta(days=1)
print(f"Ma f1ana: {ma f1ana}")
````

````python
# 📋 Mdulo json
import json

# Datos Python a JSON
persona = {
    "nombre": "Ana",
    "edad": 28,
    "hobbies": ["leer", "viajar"]
}

json_string = json.dumps(persona, indent=2, ensure_ascii=False)
print("Python a JSON:")
print(json_string)

# JSON a Python
datos = json.loads(json_string)
print(f"\nJSON a Python: {datos}")
print(f"Tipo: {type(datos)}")
````

### 📋 Formas de Importar

| Forma | Ejemplo | Cu e1ndo usar |
|-------|---------|---------------|
| `import module` | `import math` | Acceso completo: `math.sqrt()` |
| `import module as alias` | `import numpy as np` | Nombre m e1s corto |
| `from module import name` | `from math import sqrt` | Solo lo que necesitas |
| `from module import *` | `from math import *` | ⚠️ No recomendado |

**⚠️ ¿Por qu e9 evitar `import *`?**
- Contamina el namespace
- No queda claro qu e viene de d f3nde
- Puede sobrescribir funciones existentes

## 2. Crear Mdulos Personalizados

**🗂 Definici f3n:** Puedes crear tus propios mdulos organizando tu cdigo en archivos `.py`.

**📋 Estructura recomendada:**
```
mi_proyecto/
├── main.py          # Programa principal

└── config.py        # Configuraciones
```

````python
# Crear un mdulo simple dentro del notebook
# (En un proyecto real, esto ir d7a en un archivo separado)

mimodulo_code = '''
"""Mdulo de utilidades matem e1ticas."""

PI = 3.14159265359

def area_circulo(radio):
    """Calcula el  e1rea de un c d7rculo."""
    return PI * radio ** 2

def perimetro_circulo(radio):
    """Calcula el per d7metro de un c d7rculo."""
    return 2 * PI * radio

class Calculadora:
    """Una calculadora simple."""
    
    def sumar(self, a, b):
        return a + b
    
    def restar(self, a, b):
        return a - b
'''

# Escribir el mdulo a un archivo
import os
import tempfile

with tempfile.NamedTemporaryFile(mode='w', suffix='.py', delete=False) as f:
    f.write(mimodulo_code)
    module_path = f.name

print(f"Mdulo creado en: {module_path}")

# Importar y usar el mdulo
import importlib.util
spec = importlib.util.spec_from_file_location("mimodulo", module_path)
mimodulo = importlib.util.module_from_spec(spec)
spec.loader.exec_module(mimodulo)

print(f"\nPI desde mi mdulo: {mimodulo.PI}")
print(f" c1rea de c d7rculo r=5: {mimodulo.area_circulo(5):.2f}")

calc = mimodulo.Calculadora()
print(f"5 + 3 = {calc.sumar(5, 3)}")

# Limpiar
os.unlink(module_path)
````

## 3. Manejo de Excepciones

**🗂 Definicin:** Una **excepcin** es un error que ocurre durante la ejecuci f3n del programa.

**✅ Sintaxis b e1sica:**
```python
try:
    # Cdigo que puede fallar
except ErrorEspecifico:
    # Manejo del error
```

````python
# Ejemplo 1: Divisi f3n por cero
try:
    numerador = 10
    denominador = 0
    resultado = numerador / denominador
    print(f"Resultado: {resultado}")
except ZeroDivisionError:
    print("🚫 Error: 
1No se puede dividir por cero!")
````

````python
# Ejemplo 2: Manejo de entrada inv e1lida
try:
    edad = int(input("Ingresa tu edad: "))
    if edad < 0:
        raise ValueError("La edad no puede ser negativa")
    print(f"✀f8줆 Tu edad es: {edad}")
except ValueError as e:
    print(f"🚫 Entrada inv e1lida: {e}")
````

````python
# Ejemplo 3: M faltiples excepciones
try:
    # Simular diferentes errores
    opcion = 1  # Cambia este valor para probar diferentes casos
    
    if opcion == 1:
        resultado = 10 / 0
    elif opcion == 2:
        numeros[10] = 5  # IndexError
    elif opcion == 3:
        int("no es un n famero")
        
except ZeroDivisionError:
    print("Error: Divisi f3n por cero")
except IndexError:
    print("Error:  c6ndice fuera de rango")
except ValueError:
    print("Error: Valor inv e1lido")
except Exception as e:
    print(f"Error inesperado: {type(e).__name__}: {e}")
````

````python
# Ejemplo 4: else y finally

def dividir_seguro(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print("🚫 Error: Divisi f3n por cero")
        return None
    else:
        # Se ejecuta si NO hubo excepci f3n
        print(f"✅ Divisi f3n exitosa")
        return resultado
    finally:
        # Siempre se ejecuta, haya o no error
        print("📋 Bloque finally ejecutado\n")

# Caso exitoso
dividir_seguro(10, 2)

# Caso con error
dividir_seguro(10, 0)
````

## 4. Excepciones Personalizadas

**🗂 Definicin:** Puedes crear tus propias excepciones heredando de `Exception`.

````python
# Definir excepciones personalizadas
class EdadInvalidaError(Exception):
    """Excepci f3n para edad inv e1lida."""
    pass

class SaldoInsuficienteError(Exception):
    """Excepci f3n para saldo insuficiente."""
    
    def __init__(self, saldo, cantidad):
        self.saldo = saldo
        self.cantidad = cantidad
        super().__init__(f"Saldo insuficiente: tienes ${saldo}, necesitas ${cantidad}")

# Usar las excepciones personalizadas

def validar_edad(edad):
    if edad < 0:
        raise EdadInvalidaError("La edad no puede ser negativa")
    if edad > 120:
        raise EdadInvalidaError("La edad no puede ser mayor a 120")
    if edad < 18:
        raise EdadInvalidaError("Debes ser mayor de 18 a f1os")
    return True

# Ejemplo de uso
try:
    validar_edad(15)
except EdadInvalidaError as e:
    print(f"\ud83d� {e}")
````

````python
# Ejemplo: Cuenta bancaria con excepciones

class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo
    
    def retirar(self, cantidad):
        if cantidad > self.saldo:
            raise SaldoInsuficienteError(self.saldo, cantidad)
        if cantidad <= 0:
            raise ValueError("La cantidad debe ser positiva")
        self.saldo -= cantidad
        return self.saldo

# Usar la cuenta
cuenta = CuentaBancaria("Ana", 100)

try:
    print(f"Saldo inicial: ${cuenta.saldo}")
    cuenta.retir(150)  # Esto lanzar e1 excepci f3n
except SaldoInsuficienteError as e:
    print(f"\ud83d� {e}")
except ValueError as e:
    print(f"\ud83d� {e}")
````

````python
# Soluci f3n de referencia (estudiar este cdigo)

class EntradaInvalidaError(Exception):
    """Excepci f3n base para entradas inv e1lidas."""
    pass

def validar_entero_referencia(prompt, min_val=None, max_val=None):
    """Valida y retorna un entero dentro de un rango."""
    try:
        valor = int(input(prompt))
    except ValueError:
        raise EntradaInvalidaError("Debe ingresar un n famero entero")
    
    if min_val is not None and valor < min_val:
        raise EntradaInvalidaError(f"El valor debe ser al menos {min_val}")
    
    if max_val is not None and valor > max_val:
        raise EntradaInvalidaError(f"El valor debe ser m e1ximo {max_val}")
    
    return valor

def validar_email_referencia(email):
    """Valida formato de email."""
    if "@" not in email or "." not in email:
        raise EntradaInvalidaError("Email inv e1lido: debe contener @ y .")
    
    partes = email.split("@")
    if len(partes) != 2:
        raise EntradaInvalidaError("Email inv e1lido: formato incorrecto")
    
    usuario, dominio = partes
    if len(usuario) == 0 or len(dominio) == 0:
        raise EntradaInvalidaError("Email inv e1lido: usuario o dominio vac d7os")
    
    return email.lower().strip()

def validar_telefono_referencia(telefono):
    """Valida formato de tel e9fono colombiano."""
    # Eliminar espacios y guiones
    telefono = telefono.replace(" ", "").replace("-", "")
    
    # Verificar que sean solo d d7gitos
    if not telefono.isdigit():
        raise EntradaInvalidaError("Tel e9fono inv e1lido: solo debe contener n fameros")
    
    # Verificar longitud (Colombia: 10 d d7gitos m e1s indicativo)
    if len(telefono) not in [10, 12]:
        raise EntradaInvalidaError("Tel e9fono inv e1lido: debe tener 10 o 12 d d7gitos")
    
    return telefono
````
