## üìö Teor√≠a: Errores y Excepciones

### Tipos de Errores

| Tipo | Descripcion | Ejemplo |
|------|-------------|---------|
| **SyntaxError** | Error de sintaxis | `print(`hola` |
| **NameError** | Variable no definida | `print(x)` sin definir x |
| **TypeError** | Tipo incorrecto | `'2' + 2` |
| **ValueError** | Valor incorrecto | `int('abc')` |
| **IndexError** | Indice fuera de rango | `lista[10]` en lista de 5 |
| **KeyError** | Clave no existe | `dict['clave']` que no existe |
| **FileNotFoundError** | Archivo no existe | Abrir archivo inexistente |
| **ZeroDivisionError** | Division por cero | `10 / 0` |

## üõ°Ô∏è Manejo de Excepciones

```python
# ============================================
# TRY-EXCEPT BASICO
# ============================================

# Sin manejo de errores (programa se detiene)
# resultado = 10 / 0  # Esto causaria ZeroDivisionError

# Con manejo de errores
try:
    resultado = 10 / 0
except ZeroDivisionError:
    print('Error: No se puede dividir por cero')
    resultado = None

print(f'Resultado: {resultado}')
print('Programa continua...')
```

```python
# ============================================
# MULTIPLES EXCEPCIONES
# ============================================

def procesar_entrada(valor):
    try:
        numero = int(valor)
        resultado = 100 / numero
        return resultado
    except ValueError:
        print(f'Error: valor no es un numero valido')
    except ZeroDivisionError:
        print('Error: No puedes dividir por cero')
    except Exception as e:
        print(f'Error inesperado: {e}')
    return None

# Pruebas
print('Prueba 1 (numero valido):')
print(procesar_entrada('5'))

print('Prueba 2 (texto):')
print(procesar_entrada('abc'))

print('Prueba 3 (cero):')
print(procesar_entrada('0'))
```

## üìÅ Manejo de Archivos

```python
# ============================================
# ESCRIBIR Y LEER ARCHIVOS
# ============================================

# Escribir archivo
with open('ejemplo.txt', 'w', encoding='utf-8') as archivo:
    archivo.write('Linea 1: Hola mundo')
    archivo.write('Linea 2: Python es genial')

# Leer archivo
with open('ejemplo.txt', 'r', encoding='utf-8') as archivo:
    contenido = archivo.read()
    print(contenido)
```
