## ðŸš€ Tu Primera App FastAPI

````python
# ============================================
# APP BASICA FASTAPI
# ============================================

# Nota: En notebook no podemos ejecutar FastAPI directamente
# pero podemos ver el codigo y ejecutar en archivo .py aparte

CODIGO_FASTAPI = '''
from fastapi import FastAPI

# Crear instancia de la aplicacion
app = FastAPI(
    title="Mi Primera API",
    description="Una API simple con FastAPI",
    version="1.0.0"
)

# Endpoint GET basico
@app.get("/")
async def root():
    return {"message": "Â¡Hola Mundo!"}

# Endpoint con informacion
@app.get("/info")
async def info():
    return {
        "app": "Mi Primera API",
        "version": "1.0.0",
        "framework": "FastAPI"
    }

# Endpoint con parametro de ruta
@app.get("/items/{item_id}")
async def get_item(item_id: int):
    return {"item_id": item_id}

# Ejecutar con:
# uvicorn main:app --reload
'''

print(CODIGO_FASTAPI)
````

### Estructura de Rutas

````python
# ============================================
# DECORADORES DE RUTA
# ============================================

DECORADORES_EJEMPLO = '''
from fastapi import FastAPI

app = FastAPI()

# GET - Obtener datos
@app.get("/users")
async def list_users():
    return {"users": []}

# GET con parametro
@app.get("/users/{user_id}")
async def get_user(user_id: int):
    return {"user_id": user_id}

# POST - Crear datos
@app.post("/users")
async def create_user(user: dict):
    return {"created": user}

# PUT - Actualizar completamente
@app.put("/users/{user_id}")
async def update_user(user_id: int, user: dict):
    return {"updated": user_id, "data": user}

# PATCH - Actualizar parcialmente
@app.patch("/users/{user_id}")
async def patch_user(user_id: int, user: dict):
    return {"patched": user_id, "data": user}

# DELETE - Eliminar
@app.delete("/users/{user_id}")
async def delete_user(user_id: int):
    return {"deleted": user_id}
'''

print(DECORADORES_EJEMPLO)
````

## ðŸ”— Conexion con TaskFlow

La API de TaskFlow esta construida con FastAPI:

```python
# src/taskflow/api/app.py
from fastapi import FastAPI
from .routes import auth, usuarios, proyectos, tareas

app = FastAPI(
    title="TaskFlow API",
    description="API para gestion de tareas",
    version="1.0.0"
)

# Incluir routers
app.include_router(auth.router, prefix="/api/auth")
app.include_router(usuarios.router, prefix="/api/usuarios")
app.include_router(proyectos.router, prefix="/api/proyectos")
app.include_router(tareas.router, prefix="/api/tareas")
```
