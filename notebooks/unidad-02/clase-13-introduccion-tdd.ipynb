## ğŸ“š Teoria: Test Driven Development

### Â¿Que es TDD?

**TDD** (Test Driven Development) es una metodologia de desarrollo donde:
1. Escribes una prueba que falla (Red)
2. Escribes el codigo minimo para pasar la prueba (Green)
3. Refactorizas el codigo (Refactor)

### Ciclo TDD

```
    Red          Green        Refactor
  (Escribe      (Codigo       (Mejora
   test que      minimo         codigo
   falla)        para pasar)    manteniendo
                               tests verdes)
    â†‘                                    |
    â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
```

### Beneficios
- âœ… Codigo con alta cobertura de tests
- âœ… Diseno mas limpio y desacoplado
- âœ… Documentacion viva (los tests)
- âœ… Confianza al hacer cambios
- âœ… Menos bugs en produccion

## ğŸ”„ TDD en Accion

````python
# ============================================
# EJEMPLO: Funcion sumar() con TDD
# ============================================

# Paso 1: RED - Escribir test que falla
# (Imagina que escribimos esto primero)

def test_sumar():
    """Tests para la funcion sumar."""
    # Caso basico
    assert sumar(2, 3) == 5, "2 + 3 deberia ser 5"
    
    # Caso con negativos
    assert sumar(-1, 1) == 0, "-1 + 1 deberia ser 0"
    
    # Caso con cero
    assert sumar(0, 5) == 5, "0 + 5 deberia ser 5"
    
    print('âœ… Todos los tests pasan!')

# Paso 2: GREEN - Implementacion minima
def sumar(a, b):
    """Suma dos numeros."""
    return a + b

# Ejecutar tests
test_sumar()
````

````python
# ============================================
# EJEMPLO: Clase Calculadora con TDD
# ============================================

class Calculadora:
    """Calculadora basica con historial."""
    
    def __init__(self):
        self.historial = []
    
    def sumar(self, a, b):
        resultado = a + b
        self.historial.append(f'{a} + {b} = {resultado}')
        return resultado
    
    def restar(self, a, b):
        resultado = a - b
        self.historial.append(f'{a} - {b} = {resultado}')
        return resultado
    
    def multiplicar(self, a, b):
        resultado = a * b
        self.historial.append(f'{a} * {b} = {resultado}')
        return resultado
    
    def dividir(self, a, b):
        if b == 0:
            raise ValueError('No se puede dividir por cero')
        resultado = a / b
        self.historial.append(f'{a} / {b} = {resultado}')
        return resultado
    
    def obtener_historial(self):
        return self.historial.copy()
    
    def limpiar_historial(self):
        self.historial.clear()

# Tests para la calculadora
def test_calculadora():
    calc = Calculadora()
    
    # Test sumar
    assert calc.sumar(2, 3) == 5
    
    # Test restar
    assert calc.restar(5, 3) == 2
    
    # Test multiplicar
    assert calc.multiplicar(4, 3) == 12
    
    # Test dividir
    assert calc.dividir(10, 2) == 5.0
    
    # Test division por cero
    try:
        calc.dividir(10, 0)
        assert False, "Deberia haber lanzado ValueError"
    except ValueError:
        pass
    
    # Test historial
    historial = calc.obtener_historial()
    assert len(historial) == 4  # 4 operaciones exitosas
    
    # Test limpiar historial
    calc.limpiar_historial()
    assert len(calc.obtener_historial()) == 0
    
    print('âœ… Tests de Calculadora pasan!')

test_calculadora()
````
