## ðŸ’ª Ejercicios de Repaso

````python
# ============================================
# EJERCICIO INTEGRADOR: SISTEMA BIBLIOTECA
# ============================================

from abc import ABC, abstractmethod
from datetime import datetime, timedelta

# --- Clases Abstractas ---
class Material(ABC):
    """Clase abstracta para materiales de biblioteca."""
    
    def __init__(self, titulo, autor, anio):
        self.titulo = titulo
        self.autor = autor
        self.anio = anio
        self._disponible = True
    
    @property
    def disponible(self):
        return self._disponible
    
    @abstractmethod
    def dias_prestamo(self):
        """Retorna los dias de prestamo permitidos."""
        pass
    
    @abstractmethod
    def tipo_material(self):
        """Retorna el tipo de material."""
        pass
    
    def prestar(self):
        if self._disponible:
            self._disponible = False
            return True
        return False
    
    def devolver(self):
        self._disponible = True
        return True
    
    def __str__(self):
        estado = 'Disponible' if self._disponible else 'Prestado'
        return f'{self.titulo} ({self.anio}) - {estado}'

# --- Implementaciones ---
class Libro(Material):
    def __init__(self, titulo, autor, anio, isbn, genero):
        super().__init__(titulo, autor, anio)
        self.isbn = isbn
        self.genero = genero
    
    def dias_prestamo(self):
        return 14
    
    def tipo_material(self):
        return 'Libro'
    
    def __str__(self):
        return f'[Libro] {super().__str__()} - ISBN: {self.isbn}'

class Revista(Material):
    def __init__(self, titulo, autor, anio, issn, numero):
        super().__init__(titulo, autor, anio)
        self.issn = issn
        self.numero = numero
    
    def dias_prestamo(self):
        return 7
    
    def tipo_material(self):
        return 'Revista'
    
    def __str__(self):
        return f'[Revista] {super().__str__()} - N. {self.numero}'

# --- Sistema de Usuarios ---
class Usuario:
    def __init__(self, nombre, id_usuario):
        self.nombre = nombre
        self.id_usuario = id_usuario
        self.__prestamos = []
        self._limite_prestamos = 3
    
    @property
    def prestamos_activos(self):
        return len(self.__prestamos)
    
    def puede_pedir_prestamo(self):
        return self.prestamos_activos < self._limite_prestamos
    
    def agregar_prestamo(self, material):
        if not self.puede_pedir_prestamo():
            return False
        
        if material.prestar():
            fecha_devolucion = datetime.now() + timedelta(days=material.dias_prestamo())
            self.__prestamos.append({
                'material': material,
                'fecha_devolucion': fecha_devolucion
            })
            return True
        return False
    
    def devolver_material(self, material):
        for prestamo in self.__prestamos:
            if prestamo['material'] == material:
                material.devolver()
                self.__prestamos.remove(prestamo)
                return True
        return False
    
    def listar_prestamos(self):
        print(f'\nPrestamos de {self.nombre}:')
        for p in self.__prestamos:
            print(f'  - {p["material"].titulo} (Devolver: {p["fecha_devolucion"].strftime("%Y-%m-%d")})')

class Estudiante(Usuario):
    def __init__(self, nombre, id_usuario, carrera):
        super().__init__(nombre, id_usuario)
        self.carrera = carrera
        self._limite_prestamos = 5  # Mas prestamos para estudiantes

class Profesor(Usuario):
    def __init__(self, nombre, id_usuario, departamento):
        super().__init__(nombre, id_usuario)
        self.departamento = departamento
        self._limite_prestamos = 10  # Mas prestamos para profesores

# --- Demostracion ---
print('=== SISTEMA DE BIBLIOTECA ===')

# Crear materiales
libro1 = Libro('Python Clean Code', 'John Smith', 2023, '123-456', 'Programacion')
libro2 = Libro('Data Science', 'Jane Doe', 2022, '789-012', 'Ciencia de Datos')
revista1 = Revista('National Geographic', 'Varios', 2024, 'ISSN-001', 245)

# Crear usuarios
estudiante = Estudiante('Ana Garcia', 'EST-001', 'Ingenieria')
profesor = Profesor('Dr. Lopez', 'PROF-001', 'Computacion')

print(f'\nMateriales disponibles:')
print(libro1)
print(libro2)
print(revista1)

# Realizar prestamos
print(f'\n--- Prestamos ---')
print(f'{estudiante.nombre} pide {libro1.titulo}: {estudiante.agregar_prestamo(libro1)}')
print(f'{estudiante.nombre} pide {revista1.titulo}: {estudiante.agregar_prestamo(revista1)}')
print(f'{profesor.nombre} pide {libro2.titulo}: {profesor.agregar_prestamo(libro2)}')

# Listar prestamos
estudiante.listar_prestamos()
profesor.listar_prestamos()

print(f'\nEstado actual:')
print(libro1)
print(libro2)
print(revista1)
````
