## ðŸ“š Teoria: Herencia

### Â¿Que es la Herencia?

Herencia permite crear **nuevas clases basadas en clases existentes**, reutilizando codigo y estableciendo relaciones 'es-un'.

**Ejemplo:**
- Un `Estudiante` **es un** `Usuario`
- Un `Profesor` **es un** `Usuario`
- Un `Admin` **es un** `Usuario`

### Sintaxis

```python
class ClasePadre:
    pass

class ClaseHija(ClasePadre):
    pass
```

## ðŸŒ³ Herencia Simple

````python
# ============================================
# CLASE BASE (PADRE)
# ============================================

class Usuario:
    def __init__(self, nombre, email):
        self.nombre = nombre
        self.email = email
        self.activo = True
    
    def presentarse(self):
        return f'Hola, soy {self.nombre}'
    
    def desactivar(self):
        self.activo = False
        return f'{self.nombre} ha sido desactivado'

# ============================================
# CLASES DERIVADAS (HIJAS)
# ============================================

class Estudiante(Usuario):
    def __init__(self, nombre, email, carrera):
        # Llamar al constructor del padre
        super().__init__(nombre, email)
        self.carrera = carrera
        self.notas = []
    
    def agregar_nota(self, nota):
        self.notas.append(nota)
    
    def promedio(self):
        if not self.notas:
            return 0
        return sum(self.notas) / len(self.notas)
    
    # Sobrescribir metodo del padre
    def presentarse(self):
        base = super().presentarse()
        return f'{base} y estudio {self.carrera}'

class Profesor(Usuario):
    def __init__(self, nombre, email, departamento):
        super().__init__(nombre, email)
        self.departamento = departamento
        self.materias = []
    
    def asignar_materia(self, materia):
        self.materias.append(materia)
    
    def presentarse(self):
        base = super().presentarse()
        return f'{base}, profesor de {self.departamento}'

# Usar las clases
est = Estudiante('Ana Garcia', 'ana@email.com', 'Ingenieria')
prof = Profesor('Dr. Lopez', 'lopez@email.com', 'Computacion')

print(est.presentarse())
print(prof.presentarse())

est.agregar_nota(4.5)
est.agregar_nota(4.0)
print(f'Promedio de {est.nombre}: {est.promedio():.2f}')
````

## ðŸ” Funciones para Inspeccionar Herencia

````python
# ============================================
# INSPECCION DE CLASES
# ============================================

# Verificar si es instancia
print(f'est es Estudiante: {isinstance(est, Estudiante)}')
print(f'est es Usuario: {isinstance(est, Usuario)}')
print(f'est es Profesor: {isinstance(est, Profesor)}')

# Verificar si es subclase
print(f'Estudiante hereda de Usuario: {issubclass(Estudiante, Usuario)}')
print(f'Profesor hereda de Estudiante: {issubclass(Profesor, Estudiante)}')

# Ver metodos resueltos (MRO - Method Resolution Order)
print(f'\nMRO de Estudiante: {Estudiante.__mro__}')
````
