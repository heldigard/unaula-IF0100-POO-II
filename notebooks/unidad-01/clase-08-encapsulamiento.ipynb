## ðŸ“š Teoria: Encapsulamiento

### Â¿Que es el Encapsulamiento?

Encapsulamiento es el principio de **ocultar los detalles internos** de un objeto y exponer solo lo necesario.

**Beneficios:**
- âœ… Protege datos de modificaciones accidentales
- âœ… Valida datos antes de asignarlos
- âœ… Facilita el mantenimiento del codigo
- âœ… Controla el acceso a los atributos

### Convenciones en Python

| Convencion | Significado | Ejemplo |
|------------|-------------|---------|
| `atributo` | Publico | `self.nombre` |
| `_atributo` | Protegido (uso interno) | `self._saldo` |
| `__atributo` | Privado (name mangling) | `self.__password` |

## ðŸ”’ Atributos Privados y Protegidos

````python
# ============================================
# ENCAPSULAMIENTO BASICO
# ============================================

class CuentaBancaria:
    def __init__(self, titular, saldo_inicial):
        self.titular = titular          # Publico
        self._saldo = saldo_inicial     # Protegido (convencion)
        self.__pin = "1234"             # Privado (name mangling)
    
    def mostrar_info(self):
        print(f'Titular: {self.titular}')
        print(f'Saldo: ${self._saldo}')
        print(f'PIN: {self.__pin}')  # Acceso permitido desde dentro

# Crear cuenta
cuenta = CuentaBancaria('Ana Garcia', 1000)

# Acceso publico (permitido)
print(f'Titular: {cuenta.titular}')

# Acceso protegido (posible pero desaconsejado)
print(f'Saldo (protegido): {cuenta._saldo}')

# Acceso privado (dificultado por name mangling)
try:
    print(cuenta.__pin)  # Error!
except AttributeError:
    print('Error: __pin es privado')

# Pero aun es posible acceder (no recomendado)
print(f'PIN accedido via name mangling: {cuenta._CuentaBancaria__pin}')
````

## ðŸŽ¯ Properties: Getters y Setters Pythonicos

````python
# ============================================
# USANDO @PROPERTY (FORMA PYTHONICA)
# ============================================

class Producto:
    def __init__(self, nombre, precio):
        self._nombre = nombre
        self._precio = precio  # Usa el setter
    
    @property
    def nombre(self):
        """Getter para nombre."""
        return self._nombre
    
    @nombre.setter
    def nombre(self, valor):
        """Setter para nombre."""
        if not valor or len(valor) < 2:
            raise ValueError('El nombre debe tener al menos 2 caracteres')
        self._nombre = valor
    
    @property
    def precio(self):
        """Getter para precio."""
        return self._precio
    
    @precio.setter
    def precio(self, valor):
        """Setter con validacion."""
        if valor < 0:
            raise ValueError('El precio no puede ser negativo')
        self._precio = valor
    
    @property
    def precio_con_iva(self):
        """Propiedad de solo lectura (sin setter)."""
        return self._precio * 1.19

# Usar el producto
producto = Producto('Laptop', 1000)

# Acceso como atributo (pero con logica encapsulada)
print(f'Nombre: {producto.nombre}')
print(f'Precio: ${producto.precio}')
print(f'Precio con IVA: ${producto.precio_con_iva:.2f}')

# Modificar con validacion automatica
producto.precio = 1200
print(f'Nuevo precio: ${producto.precio}')

# Intentar asignar valor invalido
try:
    producto.precio = -100
except ValueError as e:
    print(f'Error: {e}')
````

## ðŸ”— Conexion con TaskFlow

En TaskFlow usamos encapsulamiento para proteger datos sensibles:

```python
class Usuario:
    def __init__(self, username, email, password):
        self.username = username
        self.email = email
        self.__password_hash = self._hash_password(password)
    
    @property
    def email(self):
        return self._email
    
    @email.setter
    def email(self, valor):
        if '@' not in valor:
            raise ValueError('Email invalido')
        self._email = valor
    
    def verificar_password(self, password):
        return self._hash_password(password) == self.__password_hash
```
